# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
df= pd.read_csv('Position_Salaries.csv')
X = df.iloc[:,-2:-1].values
y = df.iloc[:,-1].values
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X,y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures 
pol = PolynomialFeatures(degree=4)
X_pol = pol.fit_transform(X) 
lr2 = LinearRegression().fit(X_pol, y) 
```

## Visualising the Linear Regression results

```
plt.scatter(X,y)
plt.plot(lr.predict(X), color="red")
plt.title("Using Linear Regression")
plt.xlabel = "Possiiton Level"
plt.ylabel = "Salary"
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(X,y)
plt.plot(lr2.predict(X_pol), color="red")
plt.title("Using Polynomial Regression")
plt.xlabel = "Possiiton Level"
plt.ylabel = "Salary"
plt.show()
```

## Predicting a new result with Linear Regression

```
pr = lr.predict([[6.5]])
print("Slaray shoule be: ", pr)
```

## Predicting a new result with Polynomial Regression

```
pr2 = lr2.predict(pol.fit_transform([[6.5]]))
print("Salary should be: ", pr2)
```
